```python
#@markdown Librerias

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math
```

```python
#@markdown Geométrica
p =  0.3 # parametro de forma
geometrica = stats.geom(p) # Distribución
x = np.arange(geometrica.ppf(0.01),
              geometrica.ppf(0.99))
fmp = geometrica.pmf(x) # Función de Masa de Probabilidad


plt.figure(facecolor="whitesmoke")
plt.plot(x, fmp, '--',color="k",label="fmp")
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.1)

mediana = math.ceil(math.log2(2/3) / math.log2(1-p))

plt.vlines(mediana,0,max(fmp),label="Mediana", color="darkslateblue")


E=1/p
plt.vlines(E,0,max(fmp), color="red",label="E(X)")  #Media
#plt.vlines(geometrica.mean(),0,max(fmp), color="red",label="E(X)")


V=(1-p)/(p**2)
des=V**0.5
plt.fill_betweenx([0,max(fmp)], E-V,E+V, color='lightyellow', alpha=0.7,label=r"V(X)=$\sigma^2$")
#plt.fill_betweenx([0,max(fmp)], E-geometrica.var(),E+geometrica.var(), color='green', alpha=0.7,label=r"V(X)=$\sigma^2$")
plt.fill_betweenx([0,max(fmp)], E-des,E+des, color='pink', alpha=0.2, label=r"$\sigma$")
plt.title('Distribución Geométrica')
plt.ylabel('Probabilidad P(X)')
plt.xlabel('Variable aleatoria X')
plt.xlim(min(x),max(x))
plt.legend()
plt.show()
```

```python
#@markdown Geométrica
p_i=np.linspace(0.3,1,5)



plt.figure(facecolor="whitesmoke")

for i in range(len(p_i)):
  distribucion=stats.geom(p_i[i])
  x = np.arange(distribucion.ppf(0.01),distribucion.ppf(0.99))
  fmp = distribucion.pmf(x)

  #gráfica
  plt.plot(x, fmp, '--',linewidth=0.7,label=f"p={p_i[i]}")


plt.title('Función de masa de probabilidad para Distribución Geométrica con distintos p')
plt.ylabel('Probabilidad P(X)')
plt.xlabel('Variable aleatoria X')
plt.legend()
plt.grid()
plt.show()
```

```python
#@markdown Bernoulli
p =  0.5 # parametro de forma
bernoulli = stats.bernoulli(p)
x = np.arange(0, 2)
fmp = bernoulli.pmf(x) # Función de Masa de Probabilidad
fig, ax = plt.subplots(facecolor="whitesmoke")
ax.plot(x, fmp, 'ko',label="fmp")
ax.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
ax.vlines(bernoulli.mean(),0,max(fmp), color="red",label="E(X)")
plt.fill_betweenx([0,max(fmp)], bernoulli.mean()-bernoulli.var(),bernoulli.mean()+bernoulli.var(), color='lightyellow', alpha=0.7,label=r"V(X)=$\sigma^2$")
plt.fill_betweenx([0,max(fmp)], bernoulli.mean()-bernoulli.std(),bernoulli.mean()+bernoulli.std(), color='pink', alpha=0.2, label=r"$\sigma$")
plt.vlines(bernoulli.median(),0,max(fmp),linestyles="--",label="Mediana", color="darkslateblue")

plt.title('Distribución Bernoulli')
plt.ylabel('Probabilidad P(X)')
plt.xlabel('Variable aleatoria X')
plt.legend()
plt.show()
```

```python
#@markdown Binomial Negativa
from scipy.stats import nbinom
fig, ax = plt.subplots(1, 1, facecolor="whitesmoke")
n, p = 5, 0.2
mean, var, skew, kurt = nbinom.stats(n, p, moments='mvsk')
x = np.arange(nbinom.ppf(0.01, n, p),
              nbinom.ppf(0.99, n, p))
ax.plot(x, nbinom.pmf(x, n, p), '--', color="k", label='fmp')
ax.vlines(x, 0, nbinom.pmf(x, n, p),  colors='b',lw=5, alpha=0.1)
rv = nbinom(n, p)

fmp=nbinom.pmf(x, n, p)
ax.vlines(rv.mean(),0,max(fmp), color="red",label="E(X)")
plt.fill_betweenx([0,max(fmp)], rv.mean()-rv.var(),rv.mean()+rv.var(), color='lightyellow', alpha=0.7,label=r"V(X)=$\sigma^2$")
plt.fill_betweenx([0,max(fmp)], rv.mean()-rv.std(),rv.mean()+rv.std(), color='pink', alpha=0.2, label=r"$\sigma$")

mediana = (n-1)/p
plt.vlines(mediana,0,max(fmp),linestyles="--",label="Mediana", color="darkslateblue")

plt.title('Distribución Binomial Negativa')
plt.ylabel('Probabilidad P(X)')
plt.xlabel('Variable aleatoria X')

#ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,label='frozen pmf')
ax.legend(loc='best', frameon=False)
ax.set_xlim(min(x),max(x))
plt.show()
```

```python
#@markdown Hipergeométrica
M, n, N = 500, 20,300
 # parametros de forma
hipergeometrica = stats.hypergeom(M, n, N) # Distribución
x = np.arange(0, n+1)
fmp = hipergeometrica.pmf(x) # Función de Masa de Probabilidad
plt.figure(facecolor="whitesmoke")
plt.plot(x, fmp, '--',color="k",label="fmp")


plt.vlines(hipergeometrica.mean(),0,max(fmp), color="red",label="E(X)")
plt.fill_betweenx([0,max(fmp)], hipergeometrica.mean()-hipergeometrica.var(),hipergeometrica.mean()+hipergeometrica.var(), color='lightyellow', alpha=0.7,label=r"V(X)=$\sigma^2$")
plt.fill_betweenx([0,max(fmp)], hipergeometrica.mean()-hipergeometrica.std(),hipergeometrica.mean()+hipergeometrica.std(), color='pink', alpha=0.2, label=r"$\sigma$")

plt.vlines(hipergeometrica.median(),0,max(fmp),linestyles="--",label="Mediana", color="darkslateblue")


plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.1)
plt.title('Distribución Hipergeométrica')
plt.ylabel('Probabilidad P(X)')
plt.xlabel('Variable aleatoria X')
plt.legend()
plt.show()
```

```python
#@markdown Hipergeométrica
M=500
n=np.linspace(50,100,3)
N=np.linspace(100,300,2)


plt.figure(facecolor="whitesmoke")

for i in range(len(n)):
  for j in range(len(N)):
    distribucion=stats.hypergeom(M, n[i], N[j])
    x = np.arange(0, n[i]+1)
    fmp = distribucion.pmf(x)

    #gráfica
    plt.plot(x, fmp, '--',linewidth=0.7,label=f"Submuestra n={n[i]:.0f},Éxitos M={N[j]:.1f}")


plt.title('Función de masa de probabilidad para Distribución Hipergeométrica con Número total N=500')
plt.ylabel('Probabilidad P(X)')
plt.xlabel('Variable aleatoria X')
plt.legend()
plt.grid()
plt.show()
```

```python
#@markdown Log-Normal
sigma = 0.6 # parametro
lognormal = stats.lognorm(sigma)
x = np.linspace(lognormal.ppf(0.01),
                lognormal.ppf(0.99), 100)
fp = lognormal.pdf(x) # Función de Probabilidad

plt.figure(facecolor="whitesmoke")

plt.plot(x, fp,color="k",label="fdp")

distribucion=lognormal

plt.vlines(distribucion.mean(),0,max(fp), color="red",label="E(X)")
plt.fill_betweenx([0,max(fp)], distribucion.mean()-distribucion.var(),distribucion.mean()+distribucion.var(), color='lightyellow', alpha=0.7,label=r"V(X)=$\sigma^2$")
plt.fill_betweenx([0,max(fp)], distribucion.mean()-distribucion.std(),distribucion.mean()+distribucion.std(), color='pink', alpha=0.2, label=r"$\sigma$")
plt.vlines(distribucion.median(),0,max(fp),linestyles="--",label="Mediana", color="darkslateblue")

plt.title('Distribución Log-normal')
plt.ylabel('Probabilidad f(X)')
plt.xlabel('Variable aleatoria X')
plt.legend()
plt.show()
```
