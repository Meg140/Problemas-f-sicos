```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rand
```

```python
def gen_aleat(N,value):
  lista=rand.rand(N)
  cara=np.array([lista[i]<value for i in range(len(lista))])
  Ncara=sum([cara[i]==True for i in range(len(cara))])
  Nsello=sum([cara[i]==False for i in range(len(cara))])
  return Ncara,Nsello
```

```python
repeticiones=np.arange(10,8100,100)
```

```python
fcaras=np.zeros(len(repeticiones))
fsellos=np.zeros(len(repeticiones))

for i in range(len(repeticiones)):
  Ncara,Nsello=gen_aleat(repeticiones[i],0.5)
  fcaras[i]=Ncara/repeticiones[i]
  fsellos[i]=Nsello/repeticiones[i]
```

```python
plt.figure(1,facecolor="whitesmoke",figsize=(10,5))
plt.title("Figura 1. Frecuencia de ocurrencia de Cara o Sello en función del número de repeticiones.")
plt.plot(repeticiones,fcaras,label=r"$f_{cara}$",color="yellowgreen")
plt.plot(repeticiones,fsellos,label=r"$f_{sello}$",color="mediumorchid")
plt.hlines(0.5,0,max(repeticiones),color="k")
plt.xlabel(r"Número de repeticiones $N$")
plt.ylabel(r"Frecuencia $f$")
plt.legend()
plt.grid()
```

```python
fcaras1=np.zeros(len(repeticiones))
fsellos1=np.zeros(len(repeticiones))

for i in range(len(repeticiones)):
  Ncara,Nsello=gen_aleat(repeticiones[i],0.8)
  fcaras1[i]=Ncara/repeticiones[i]
  fsellos1[i]=Nsello/repeticiones[i]
```

```python
plt.figure(2,facecolor="whitesmoke",figsize=(10,5))
plt.title("Figura 2. Frecuencia con asignación de probabilidad 80% para Cara y 20% para Sello.")
plt.plot(repeticiones,fcaras1,label=r"$f_{cara}$",color="yellowgreen")
plt.plot(repeticiones,fsellos1,label=r"$f_{sello}$",color="mediumorchid")
plt.hlines(0.8,0,max(repeticiones),color="k")
plt.hlines(0.2,0,max(repeticiones),color="k")
plt.xlabel("Número de repeticiones")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid()
```
